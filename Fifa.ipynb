<p>Whereas <code class="docutils literal notranslate"><span class="pre">read_*</span></code> functions are used to read data to pandas, the
<code class="docutils literal notranslate"><span class="pre">to_*</span></code> methods are used to store data. The <a class="reference internal" href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_excel.html#pandas.DataFrame.to_excel" title="pandas.DataFrame.to_excel"><code class="xref py py-meth docutils literal notranslate"><span class="pre">to_excel()</span></code></a> method stores
the data as an excel file. In the example here, the <code class="docutils literal notranslate"><span class="pre">sheet_name</span></code> is
named <em>passengers</em> instead of the default <em>Sheet1</em>. By setting
<code class="docutils literal notranslate"><span class="pre">index=False</span></code> the row index labels are not saved in the spreadsheet.</p>
<hr class="task-bullet-end" style="height:1em;background:transparent;border:none;border-bottom:2px solid #130654;border-radius:1em;"/>
